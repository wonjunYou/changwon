```
location = df['관할행정동'].str.lstrip()
```

```
location
```

```
location_list = list(location)
```

```
location_dict = {}
prev_location=""
count = 1

for i in range(len(location_list)):
    if i==0:
        prev_location="의창구 동읍 "
    else:
        if prev_location == location_list[i]:
            count = count + 1
        else:
            location_dict[prev_location] = count
            count=1
        prev_location = location_list[i]
        
location_dict[prev_location] = count
```

```
location_dict
```

```
#dictionary to dataFrame
total_df = pd.DataFrame.from_dict(location_dict,orient='index')
```

```
total_df.columns=['노인정 수']
total_df['동읍면'] = total_df.index
total_df = total_df[['동읍면','노인정 수']]
total_df.reset_index(drop=True)
```
