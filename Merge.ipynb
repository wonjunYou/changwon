```
df1=pd.read_csv('노인 인구수.csv')
df2=pd.read_csv('문화생활기관.csv')
df3=pd.read_csv('지역별 노인정 수.csv')
df4=pd.read_csv('무료급식소.csv')
df5=pd.read_excel('버스 정보 안내기 수.xlsx')
```

```
from functools import reduce


dfs = [df1, df2, df3, df4, df5]

df= reduce(lambda  left,right: pd.merge(left,right,on=['구','동'], how='outer'), dfs)
df
```
